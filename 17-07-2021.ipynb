```python
# pip install pandas
```

```python
import pandas as pd
```

```python
dir(pd)
```

```python
data=pd.Series()
print(data)
```

```python
data=pd.Series([1,2,3,4,"python",56.2])
print(data)
```

## Matplotlib

```python
#pip install matplotlib
```

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6]
y=[10,45,3,12,67,4]
y1=[2,45,10,56,34,56]
plt.title("sample graph")
plt.xlabel("x---axis")
plt.ylabel("y----axis")
plt.grid()
plt.plot(x,y,'--g',label="first graph")
plt.plot(x,y1,label="second graph")
plt.legend()
plt.show()
```

```python
plt.bar(x,y)
```

```python
plt.scatter(x,y)
```
